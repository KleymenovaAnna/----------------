```python
from PIL import Image, ImageDraw, ImageFont
import os

def create_image_preview(image_list, direction='vertical', cells=3, cell_size=(100, 100), max_rows=None, with_labels=True):
    images = [Image.open(img) for img in image_list]
    preview_image = None
    width = cells * cell_size[0]
    height = ((len(images) - 1) // cells + 1) * (cell_size[1] + 40)

    if direction == 'horizontal':
        preview_image = Image.new('RGB', (width, height), color='white')
        x_offset = 0
        y_offset = 0
        for img in images:
            img.thumbnail(cell_size)
            preview_image.paste(img, (x_offset, y_offset))

            if with_labels:
                draw = ImageDraw.Draw(preview_image)
                font = ImageFont.truetype("arial.ttf", 12)
                size_text = f"Size: {img.size}"
                format_text = f"Format: {os.path.splitext(image_list[images.index(img)])[-1][1:]}"
                draw.text((x_offset, y_offset + cell_size[1]), size_text, font=font, fill='black')
                draw.text((x_offset, y_offset + cell_size[1] + 20), format_text, font=font, fill='black')

            x_offset += cell_size[0]
            if x_offset >= width:
                x_offset = 0
                y_offset += cell_size[1] + 40
    else:
        preview_image = Image.new('RGB', (width, height), color='white')
        x_offset = 0
        y_offset = 0
        for img in images:
            img.thumbnail(cell_size)
            preview_image.paste(img, (x_offset, y_offset))

            if with_labels:
                draw = ImageDraw.Draw(preview_image)
                font = ImageFont.truetype("arial.ttf", 12)
                size_text = f"Size: {img.size}"
                format_text = f"Format: {os.path.splitext(image_list[images.index(img)])[-1][1:]}"
                draw.text((x_offset, y_offset + cell_size[1]), size_text, font=font, fill='black')
                draw.text((x_offset, y_offset + cell_size[1] + 20), format_text, font=font, fill='black')

    return preview_image
```

```python
# Пример использования функции
image_list = ['C:\\prog\photo_2024-06-08_18-19-50.jpg',r'C:\\prog\photo_2024-06-08_18-20-00.jpg',r'C:\\prog\photo_2024-06-08_18-20-16.jpg',r'C:\\prog\photo_2024-06-08_18-20-31.jpg',r'C:\\prog\photo_2024-06-08_18-20-36.jpg',r'C:\\prog\photo_2024-06-08_18-20-46.jpg',r'C:\\prog\photo_2024-06-08_18-21-04.jpg',r'C:\\prog\photo_2024-06-08_18-21-10.jpg',r'C:\\prog\photo_2024-06-08_18-21-19.jpg',r'C:\\prog\photo_2024-06-08_18-21-24.jpg',r'C:\\prog\photo_2024-06-08_18-21-30.jpg',r'C:\\prog\photo_2024-06-08_18-21-35.jpg']
result_image = create_image_preview(image_list, direction='horizontal', cells=3, cell_size=(100, 100), with_labels=True)
display(result_image)
```
