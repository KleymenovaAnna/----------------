```python
from PIL import Image

def image_to_ascii_squares(image_path):
    # Используем средний квадрат Unicode в качестве символа для ASCII-арта
    square_char = '▒'

    # Загружаем изображение и преобразуем в оттенки серого
    image = Image.open(image_path).convert('L')

    # Определяем размеры консоли
    console_width = 80
    aspect_ratio = image.width / image.height
    console_height = int(console_width / aspect_ratio)

    # Масштабируем изображение
    resized_image = image.resize((console_width, console_height))

    # Преобразуем изображение в строку квадратов
    ascii_str = ''
    for y in range(console_height):
        for x in range(console_width):
            brightness = resized_image.getpixel((x, y))
            # Используем квадраты для всех символов
            ascii_str += square_char if brightness < 128 else ' '
        ascii_str += '\n'

    return ascii_str
```

```python

image_path = r'photo\photo_2024-06-08_18-19-50.jpg'
# Получаем ASCII-арт и выводим его
ascii_art = image_to_ascii_squares(image_path)
print(ascii_art)
```
